```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('spam.csv', encoding='ISO-8859-1')
```

```python
df
```

```python
df = df[['v1', 'v2']]
df.columns = ['label', 'message']
```

```python
df
```

```python
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
```

```python
## remove the special char, lower-case the char, remove stopwords, stemming:
corpus = []
for i in range(0, len(df)):
    review = re.sub('[^a-zA-Z]',' ', df['message'][i]) ##remove special ch
    review = review.lower() ## lower case
    review = review.split() ## convert to list
    review = [stemmer.stem(word) for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
## BOW model -> binary BOW(only 0 & 1):
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=1500, binary=True)
```

```python
X = cv.fit_transform(corpus).toarray()
```

```python
## N-grams implementation:
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=1500, binary=True, ngram_range=(1,2))
X = cv.fit_transform(corpus).toarray()
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
corpus = []
for i in range(0, len(df)):
    review = re.sub('[^a-zA-Z]',' ', df['message'][i]) ##remove special ch
    review = review.lower() ## lower case
    review = review.split() ## convert to list
    review = [lemmatizer.lemmatize(word) for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
tfidf = TfidfVectorizer(max_features=100)
```
